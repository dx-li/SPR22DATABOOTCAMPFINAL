```python
import numpy as np
import pandas as pd
```

```python
df_latam = pd.read_csv("PANELLATAM.csv")
df_latam.head()
```

```python
df_latam.columns =["COUNTRY", 'DATE', 'CO2E', 'CHEPC', 'IHB3', 'LBB', "MRU5", 'PUSMDWS', 'PUNDER']
```

```python
df_latam['DATE'] = pd.to_datetime(df_latam['DATE'], format = '%Y')
```

```python
for i in ['CO2E', 'CHEPC', 'IHB3', "MRU5", 'PUSMDWS', 'PUNDER']:
    label = f'log_{i}'
    df_latam[label] = np.log(df_latam[i])
```

```python
from plotnine import *
```

```python
(ggplot(df_latam, aes("CO2E", "MRU5", color = 'COUNTRY')) 
    + geom_point(alpha = 0.5) 
    + theme_classic()
    + theme(legend_position = 'none')
    + labs(title = "Latin America"))
```

```python
(ggplot(df_latam, aes("CO2E", "MRU5", color = 'COUNTRY')) 
    + geom_point(alpha = 0.5) 
    + facet_wrap("~COUNTRY")
    + theme_classic()
    + theme(legend_position = 'none')
    + labs(title = "Latin America"))
```

```python
df_latam_indexed = df_latam.set_index(["COUNTRY", "DATE"])
```

```python
df_latam_indexed.head()
```

```python
from linearmodels import PanelOLS
import statsmodels.api as sm
exog = sm.add_constant(df_latam_indexed[['PUSMDWS']])
model = PanelOLS(df_latam_indexed['MRU5'], exog, entity_effects=False, 
                 time_effects=False).fit(cov_type='clustered', cluster_entity = True, cluster_time = False)
print(model)
```
